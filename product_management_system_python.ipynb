```python
## Project Overview

# This project implements a Product Management System using Object-Oriented Programming in Python.

### Features implemented:
# - Data validation using setters
# - Encapsulation with private attributes
# - Stock management (add/remove)
# - Discount application
# - Total stock value calculation
# - Product information display


class Produto:
  def __init__(self, codigo, nome, preco, quantidade_estoque):
    self.set_codigo(codigo)
    self.set_nome(nome)
    self.set_preco(preco)
    self.set_quantidade_estoque(quantidade_estoque)

#setter Codigo
  def set_codigo(self, codigo):
    if isinstance(codigo, str) and len(codigo) > 0:
      self._codigo = codigo
    else:
      raise ValueError(f"Código deve ser maior que zero {self._codigo}")
# Getter
  def get_codigo(self):
    return self._codigo

#setter nome
  def set_nome(self, nome):
    if isinstance(nome, str) and len(nome) > 0:
      self._nome = nome
    else:
      raise ValueError(f"Nome inválido {self._codigo}")
# Getter
  def get_nome(self):
    return self._nome

    #setter preco
  def set_preco(self, preco):
    if isinstance(preco,(int,float)) and preco > 0:
      self._preco = preco
    else:
      raise ValueError(f"Preco inválido {self._codigo}")
# Getter
  def get_preco(self):
    return self._preco

#setter QTD
  def set_quantidade_estoque(self, quantidade_estoque):
    if isinstance(quantidade_estoque, int) and quantidade_estoque > 0:
      self._quantidade_estoque = quantidade_estoque
    else:
      raise ValueError(f"quantidade_estoque inválido {self._codigo}")
# Getter
  def get_quantidade_estoque(self):
    return self._quantidade_estoque

# adicionar estoque
  def adicionar_estoque(self, adc_estoque):
    self._quantidade_estoque = self._quantidade_estoque + adc_estoque
    print(f"Você adicionou {adc_estoque} peça(s), agora seu estoque total é: {self._quantidade_estoque}")

# remover estoque
  def remover_estoque(self, rem_estoque):
    if isinstance(rem_estoque, int) and rem_estoque > 0:
      if rem_estoque <= self._quantidade_estoque:
        self._quantidade_estoque -= rem_estoque
        print(f"Você removeu {rem_estoque} peça(s), agora seu estoque total é: {self._quantidade_estoque}")
        return True
      else:
        print("Quantidade em estoque insuficiente")
        return False
    print(f"vc nao pode remover {rem_estoque}")
    return False

# valor total estoque
  def calcular_valor_total_estoque(self):
    return self._quantidade_estoque * self._preco

# aplicar desconto
  def aplicar_desconto(self, percentual):
    if isinstance (percentual, int) and percentual > 0 and percentual <= 15:
      self._preco -= self._preco * (percentual / 100)
      print(f"Você aplicou um desconto de {percentual}%, novo preço {self._preco:.2f}, produto {self._codigo}")

#exibir infor
  def exibir_info(self):
    print("=" * 40)
    print(f"Código: {self._codigo}")
    print(f"Nome: {self._nome}")
    print(f"Preço: {self._preco:.2f}")
    print(f"Quantidade em estoque: {self._quantidade_estoque}")
    print("=" * 40)


produto = Produto(codigo = "XBUTD", nome = "Macbook", preco = 12999, quantidade_estoque = 10)

print(produto.get_quantidade_estoque())
produto.adicionar_estoque(12)
print(produto.get_quantidade_estoque())
produto.remover_estoque(10)
print(produto.get_quantidade_estoque())
print(produto.calcular_valor_total_estoque())
produto.aplicar_desconto(10)
produto.exibir_info()


```

```python
# Product management system with product registration, search, listing,
# total inventory value calculation, and summary reporting.


class GestaoProdutos:
  def __init__(self):
    self._produtos = {}

  def adicionar_produto(self, produto):
    if isinstance(produto, Produto): #produto adicionado é uma instancia da classe criada anteriormente
      codigo = produto.get_codigo()
      if codigo not in self._produtos:
        self._produtos[codigo] = produto
        print(f"Produto {codigo} adicionado com sucesso")
      else:
        print(f"Produto {codigo} já existe")

  def buscar_produto(self, codigo):
    return self._produtos.get(codigo)

  def listar_produtos(self):
    if self._produtos:
      print("\n === Lista de Produtos ===")
      for produto in self._produtos.values():
          produto.exibir_info()
    else:
      print("Nenhum produto encontrado")


gestao = GestaoProdutos()

gestao.adicionar_produto(Produto("PR1", "Notebook", 2500.00, 10))
gestao.adicionar_produto(Produto("PR2", "Mouse", 500.00, 11))
gestao.adicionar_produto(Produto("PR3", "Chinelo", 50.00, 9))

produto.exibir_info()

gestao.listar_produtos()

gestao.buscar_produto("PR1").aplicar_desconto(10)
gestao.buscar_produto("PR1").exibir_info()
```
