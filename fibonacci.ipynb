```python
### Description - This project calculates the Fibonacci sequence using different approaches.


fibonacci = [0,1]

limite_termos = 10
limite_valor = 100


# Fibonacci based on "for"

fibonacci = [0,1]

for i in range(2, limite_termos):
  proximo = fibonacci[i-1] + fibonacci[i-2]
  fibonacci.append(proximo)


# Fibonacci based on "while"
fibonacci_while = [0,1]
indice = 2

while True:
  proximo = fibonacci_while[indice-1] + fibonacci_while[indice-2]
  if proximo > limite_valor:
    break

  fibonacci_while.append(proximo)
  indice = indice + 1


text = f"""
{"==="*10}
SEQUÊNCIA DE FIBONACCI
Limite de termos: {limite_termos}
Limite de valor: {limite_valor}

Fibonacci based on "for": {fibonacci}
Fibonacci based on "while": {fibonacci_while}

{"==="*10}
"""

print(text)
```
